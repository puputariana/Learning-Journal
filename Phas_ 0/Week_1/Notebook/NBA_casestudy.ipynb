## grouping and aggregating

```python
df.sample(5)
```

```python
gr = df.groupby('game_result')

gr
```

```python
gr.get_group('L')
```

```python
df.groupby('game_result')[['pts']].sum()
```

```python
t#bracket pertama grouping. bracket kedua dan selanjutnya aggregasi
df.groupby('game_result')['pts'].mean()
```

```python
df[
    (df['fran_id']=='Spurs') &
    (df['year_id']>2010) 
].groupby(['year_id','game_result'])['game_id'].count()
```

```python
df[
    (df['fran_id']=='Spurs') &
    (df['year_id']>2010) 
].groupby(['year_id','game_result']).agg({'pts': 'mean', 'game_id':'count'})
```
