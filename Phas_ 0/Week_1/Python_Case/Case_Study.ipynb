```python
missing_values= ['n/a', 'na', '--']
en = pd.read_csv("property_data.csv", na_values = missing_values)
```

```python
en
```

```python
cnt=0
for row in en['OWN_OCCUPIED']:
    try:
        int(row)
        en.loc[cnt, 'OWN_OCCUPIED']=np.nan #for every nyeleneh value
    except ValueError:
        pass
    cnt+=1
```

```python
en.head(9)
```
